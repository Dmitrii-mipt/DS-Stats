Пусть $X_1, ..., X_n \sim Bern(p)$ &mdash; проведенные наблюдения, где $X_i = 1$, если в $i$-м испытании случилась неудача после того, как черный кот перебежал дорогу, а $p$ &mdash; неизвестная вероятность такого события. 
Вы хотите проверить гипотезу $\mathsf{H}_0\colon p=1/2$ против альтернативы $\mathsf{H}_1\colon p>1/2$. Нулевая гипотеза означает отсутствие связи между черным котом и неудачей, первая &mdash; что неудача происходит чаще если черный кот перебегает дорогу.

На семинаре получено, что $S = \left\{T(x) \geqslant c_\alpha\right\}$, где $T(X) = \sum\limits_{i=1}^n X_i$, является равномерно наиболее мощным критерием для проверки этих гипотез.

В каждом случае найдите значение $c_\alpha$, а также выведите результат процедуры проверки гипотез 0/1 &mdash; отвергается ли $\mathsf{H}_0$ (выводите 1, если отвергается).
Оформите это в виде таблицы, можно через `pandas.DataFrame`.

*Замечание:* Пользуйтесь функциями из `scipy.stats`.

```python
import numpy as np
import pandas as pd
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5, palette='Set2')
```

```python
alpha = 0.05
n = [10, 20, 50, 100, 500]
count = 25
p = 0.5

for i in range(len(n)):
    column = [sps.binom(n=n[i], p=p).ppf(1- alpha) + 1]
    sample = sps.binom(n=n[i], p=p).rvs(size=count)
    T = (sample >= column[0]).astype(int)
    [column.append(i) for i in T]
    reject = sum(T) / count
    column.append(reject)
    
    if i == 0:
        data = pd.DataFrame(np.array(column), columns=['n = {}'.format(n[i])], index=['$c_{\\alpha}$'] + \
                           list(np.arange(1, count + 1)) + ['reject_freq'])
        
    else: 
        data_c = pd.DataFrame(np.array(column), columns=['n = {}'.format(n[i])], index=['$c_{\\alpha}$'] + \
                           list(np.arange(1, count + 1)) + ['reject_freq'])
        data = pd.concat([data, data_c], axis=1)
data
```

```python
print('Средняя частотность ошибки:', data.iloc[count + 1].mean())
```

Теперь для некоторых $n$ постройте графики функции мощности критерия при $1/2 < p < 1$ и уровне значимости 0.05.

```python
alpha = 0.05
n = np.array([10, 20, 50, 100, 500])
p_0 = 0.5
p = np.linspace(0.5 + 0.0001, 1 - 0.0001, 10**4)
c = sps.binom(n=n.reshape(-1, 1), p=p_0).ppf(1- alpha) + 1
betta = sps.binom(n=n.reshape(-1, 1), p=p).sf(c)

plt.figure(figsize=(15, 8))
for i in range(len(n)):
    plt.plot(p, betta[i], label='n = {}'.format(n[i]))
    
plt.xlabel('p', fontsize=15)
plt.ylabel('$\\beta_S(p)$', fontsize=15)
plt.title('График зависимости мощности критерия от веротяности p', fontsize=15, y=1.02)
plt.legend();
```

**Вывод:** На графике видно, что чем больше $n$, тем быстрее мощность сходится к 1, что и логично из теории. Также мы знаем, что при $p = 0.5$ значение мощности равно значению ошибки 1 рода и мы видим, что она действительно $\leq \alpha$.

Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем не редко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью семплирования по методу Монте-Карло.

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените реальный уровень значимости** критерия для размеров выборки $1 \leqslant n \leqslant 50$ и визуализируйте полученную зависимость.

```python
alpha = 0.05
n = 50
d = 10**3
theta = 1
sample = sps.poisson(mu=theta).rvs(size=(d, n))
sample_mean = sample.cumsum(axis=1) / (np.arange(n) + 1)
q = sps.norm().ppf(1 - alpha / 2)
reject = (np.sqrt(np.arange(n) + 1) * np.abs(sample_mean - theta) / np.sqrt(sample_mean) >= q).mean(axis=0)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(np.arange(n) + 1, reject)
plt.xlabel('n')
plt.ylabel('Реальный уровень значимости')
plt.title('График зависимости реального уровня значимости от количества элементов в выборке', y=1.02);
```

**Вывод:** На графике видно, что при малых значениях $n$ график ведет себя плохо, так как реальный уровень значимости больше $\alpha$, причем значительно. При увеличении же $n$ значение реального уровня значимости уменьшается и колеблется в небольшой окрестности $\alpha$.

**2.** Для значений $n \in \{5, 15, 50\}$ **оцените функцию мощности критерия** для значений $\theta \in (0, 5)$ по сетке с шагом 0.01 по методу Монте-Карло. Визуализируйте полученные зависимости. Посчитайте также мощность критерия по асимптотической формуле, полученной на лекции, и сравните с оценкой мощности методом Монте-Карло. Учтите, что асимптотическая формула содержит $n$, поэтому для нее тоже должно быть три зависимости.

```python
n = [5, 15, 50]
max_size = 50
d = 10**3
theta = np.arange(0 + 0.01, 5 - 0.01, 0.01)
sample = []
for i in theta:
    sample.append(sps.poisson(mu=i).rvs(size=(max_size, d)))
sample = np.array(sample)

sample_mean = []
for i in n:
    sample_mean.append(np.mean(sample[:][:i][:], axis=1))
sample_mean = np.array(sample_mean)

w = (np.sqrt(n).reshape(-1, 1) * (theta - 1)).reshape(len(n), len(theta), 1) / np.sqrt(sample_mean)

q = sps.norm().ppf(1 - alpha / 2)
reject = (np.sqrt(n).reshape(-1, 1, 1) * np.abs(sample_mean - 1) / np.sqrt(sample_mean) >= q).mean(axis=2)

betta = (sps.norm().sf(sps.norm().ppf(1 - alpha / 2) - w) + sps.norm().cdf(sps.norm().ppf(alpha / 2) - w)).mean(axis=2)
```

```python
plt.figure(figsize=(20, 8))
plt.plot(theta, reject[0], label='Метод монте карло при n = 5', color='orange')
plt.plot(theta, betta[0], linestyle=':', lw=3, label = 'Асимптотически при n = 5', color='orange')

plt.plot(theta, reject[1], label='Метод монте карло при n = 15', color='red')
plt.plot(theta, betta[1], linestyle=':', lw=3, label = 'Асимптотически при n = 15', color='red')

plt.plot(theta, reject[2], label='Метод монте карло при n = 50', color='green')
plt.plot(theta, betta[2], linestyle=':', lw=3, label = 'Асимптотически при n = 50', color='green')

plt.xlabel('$\\theta$')
plt.ylabel('$\\beta_S(\\theta)$', fontsize=15)
plt.title('Сравнение Асимптотического подхода и метода Монте Карло', fontsize=20, y=1.02)
plt.legend();
```

**Вывод:** На графике видно, что при небольших n графики существенно различаются. Также видно, что чем больше n, тем больше совпадают графики при двух подходах, что и логично. Также видно, что чем больше n, тем быстрее мощность сходится к 1.

**3.** Пусть $n=50$. **Сравните мощности критериев**, соответствующих двусторонней, левосторонней и правосторонней альтернативам.

```python
n = 50
d = 10**3
theta = np.arange(0 + 0.01, 5 - 0.01, 0.01)

sample = []
for i in theta:
    sample.append(sps.poisson(mu=i).rvs(size=(n, d)))
sample = np.array(sample)

sample_mean = sample.mean(axis=1)
reject = (np.sqrt(n) * np.abs(sample_mean - 1) / np.sqrt(sample_mean) >= q).mean(axis=-1)
w = (np.sqrt(n).reshape(-1, 1) * (theta - 1)) / np.sqrt(sample_mean)
betta = (sps.norm().sf(sps.norm().ppf(1 - alpha / 2) - w) + sps.norm().cdf(sps.norm().ppf(alpha / 2) - w)).mean(axis=-1)

reject_left = (np.sqrt(n) * (sample_mean - 1) / np.sqrt(sample_mean) <= sps.norm().ppf(alpha)).mean(axis=-1)
betta_left = (sps.norm().cdf(sps.norm().ppf(alpha) - w)).mean(axis=-1)

reject_right = (np.sqrt(n) * (sample_mean - 1) / np.sqrt(sample_mean) >= sps.norm().ppf(1 - alpha)).mean(axis=-1)
betta_right = (sps.norm().sf(sps.norm().ppf(1 - alpha) - w)).mean(axis=-1)
```

```python
plt.figure(figsize=(20, 8))
plt.plot(theta, reject, label='Метод монте карло при двусторонней', color='orange')
plt.plot(theta, betta, linestyle=':', lw=3, label = 'Асимптотически при двусторонней', color='orange')

plt.plot(theta, reject_left, label='Метод монте карло при левосторонней', color='red')
plt.plot(theta, betta_left, linestyle=':', lw=3, label = 'Асимптотически при левосторонней', color='red')

plt.plot(theta, reject_right, label='Метод монте карло при правосторонней', color='green')
plt.plot(theta, betta_right, linestyle=':', lw=3, label = 'Асимптотически при правосторонней', color='green')

plt.xlabel('$\\theta$')
plt.ylabel('$\\beta_S(\\theta)$', fontsize=15)
plt.title('Сравнение Асимптотического подхода и метода Монте Карло при разных альтернативах', fontsize=20, y=1.02)
plt.legend();
```

**Вывод:** Как и говорилось на лекции, для правосторонней альтернативы наиболее мощный ее критерийб аналогично для левосторонней альтернативы. Также видно, что для двусторонней альтернативы мощность меньше, чем при односторонних, но все равно она "хорошая".
